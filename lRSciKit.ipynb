```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dfx = pd.read_csv("./linearX.csv")
dfy = pd.read_csv("./linearY.csv")

x = dfx.values
y = dfy.values
print(x)
print(y)
print(x.shape)
print(y.shape)
```

```python
x=x.reshape((-1,1))
print(x)
y=y.reshape((-1, 1))
print(y)
```

```python
plt.scatter(x,y)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(x,y)
```

```python
output = model.predict(x)
```

```python
bias = model.intercept_
coff = model.coef_
print(bias)
print(coff)
```

```python
model.score(x,y)
```

```python
plt.scatter(x,y, color='red')
plt.plot(x, output, color='black')
```

```python
from sklearn.metrics import mean_squared_error, r2_score
print(r2_score(y,output))
```
